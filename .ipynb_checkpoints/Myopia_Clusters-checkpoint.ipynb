```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
from sklearn.manifold import TSNE
import os
os.environ["OMP_NUM_THREADS"] = '1'
```

# Preparing Data

```python
#Reading CSV
file_path = Path("myopia.csv")
df = pd.read_csv(file_path)
df.head()
```

```python
labels = df["MYOPIC"]
```

```python
#Dropping "MYOPIC" column from DataFrame
X = df.drop(columns= ["MYOPIC"])
X.head()
```

```python
# Standarizing data with StandarScaler
scalar = StandardScaler()
scaled_X = scalar.fit_transform(X)
print(scaled_X[0:5])
```

# Apply Dimensionality Reduction

## Performing PCA

```python
# Initialize PCA model
pca = PCA(n_components=0.9)

# Get two principal components for the iris data.
pca_X = pca.fit_transform(scaled_X)

df_pca = pd.DataFrame(data=pca_X)
df_pca.head()
```

```python
# Viewing the number of features reduced after applying PCA (went from 14 to 10)
df_pca.shape
```

```python
# Fetch the explained variance (above 0.80)
pca.explained_variance_ratio_.sum()
```

## Running t-SNE on the principal components

```python
# Run t-SNE
tsne = TSNE(learning_rate=50, perplexity=50, random_state= 1)
tsne_features = tsne.fit_transform(df_pca)
```

```python
# Viewing the number of features reduced by t-SNE
tsne_features.shape
```

```python
#Creating a tsne_df to plot scatter plots
tsne_df =pd.DataFrame(data=tsne_features, columns=("X","Y"))
tsne_df.head()
```

## Plotting t-SNE features as scatter plots to identify clusters visually

```python
plt.scatter(tsne_df["X"], tsne_df["Y"] )
plt.show()
```

```python
plt.scatter(tsne_df["X"], tsne_df["Y"], c=labels)
plt.show()
```

## Analysis Based Scatter Plots

When we create an initail scatter plot, plotting the reduced features by t-SNE, visually we can idententiy around 4 clusters. However when we add the Myopia data as labels to the same scattter plot we can see there is a bit of over lap in the previously identified clusters. However we can see the yellow dots seem to be clustered more in upper x and y values.

# Perform a Cluster Analysis with K-means

## Creating an Elbow Plot

```python
inertia = []
#Setting range of k from 1 to 10 including both
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(tsne_features)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
# Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve')
plt.show()
```

## Elbow Curve

The Elbow Curve produced is not the best looking and easy to read, however we can get an idea of the optimal number of clusters and work further to see what would be the best number of clusters.

I feel the it is fair to say that the elbow would be at 4 as after 4 the plot seems to be decrease fairly linearly.

To examine further lets do k-means analysis using k as 3, 4 and 5 to see what the perfect number of clusters in which we can group the patients.


## KMeans, k=3

```python
# Initializing model with K = 3
model = KMeans(n_clusters=3, random_state=0)

# Fitting model and making predictions
model.fit(tsne_df)
predictions = model.predict(tsne_df)

#Adding predicted class to dataframe
tsne_df['class'] = model.labels_
tsne_df.head()
```

```python
#Creating a Scatter plot, colored per predicted class
plt.scatter(tsne_df["X"], tsne_df["Y"], c=tsne_df["class"])
plt.show()
```

## KMeans, k=4

```python
# Initializing model with K = 4
model = KMeans(n_clusters=4, random_state=0)

# Fitting model and making predictions
model.fit(tsne_df)
predictions = model.predict(tsne_df)

#Adding predicted class to dataframe
tsne_df['class'] = model.labels_
tsne_df.head()
```

```python
#Creating a Scatter plot, colored per predicted class
plt.scatter(tsne_df["X"], tsne_df["Y"], c=tsne_df["class"])
plt.show()
```

## KMeans, k=5

```python
# Initializing model with K = 5
model = KMeans(n_clusters=5, random_state=0)

# Fitting model and making predictions
model.fit(tsne_df)
predictions = model.predict(tsne_df)

#Adding predicted class to dataframe
tsne_df['class'] = model.labels_
tsne_df.head()
```

```python
#Creating a Scatter plot, colored per predicted class
plt.scatter(tsne_df["X"], tsne_df["Y"], c=tsne_df["class"])
plt.show()
```
